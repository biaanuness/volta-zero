```python


def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int, tipo_dado: str):
  coluna = []
  with open(nome_arquivo, 'r', 'utf8') as fp:
    fp.readline() # pular cabeçalho
    linha = fp.readline()
    while linha:
            linha = linha.strip() # remove espaços em branco e "\n" das pontas
            partes = linha.split(',') # separa valores em um lista pelas virgulas

            valor = partes[indice_coluna]

            if tipo_dado == 'int':
                valor = int(valor)
            elif tipo_dado == 'float':
                valor = float(valor)
            else:
                valor = str(valor)

            coluna.append(valor)
            linha = fp.readline()

    return coluna
      

```

```python
from functools import reduce 

def processar_faturas(arquivo: str) -> float:
    faturas = []

    with open(file=arquivo, mode='r', encoding='utf8') as arquivo:
        linha = arquivo.readline()
        linha = arquivo.readline()
        while linha:
            fatura = float(linha.strip().split(sep=",")[-3])
            faturas.append(fatura)
            linha = arquivo.readline()
     
    total_pagar = reduce(lambda x, y: x + y, faturas)
    total_pagar = round(total_pagar, 2)

    return total_pagar 
```

```python
total_pagar = processar_faturas(arquivo='./telecom.csv')
print(total_pagar)
```

```python
def stats_dados_faltantes(df:pd.DataFrame) -> None:
    stats_dados_faltantes = []
    for col in df.columns:
        if df[col].isna().any():
            qtd, _ = df[df[col].isna()].shape
            total, _ =df.shape 
            dict_dados_faltantes ={col: {'quantidade': qtd, "porcentagem":round(100 * qtd/total, 2)}}
            stats_dados_faltantes.append(dict_dados_faltantes)
    for stat in stats_dados_faltantes:
        print(stat)
```
